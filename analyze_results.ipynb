```python
%matplotlib inline

import os
import pandas as pd
from pyparsing import*
from pygazeanalyser.idfreader import read_idf
import numpy
import matplotlib
from matplotlib import pyplot, image
from scipy.misc import imread
import matplotlib.cbook as cbook
```

```python
data=read_idf('Timofey_test_eyetracker result.txt', 'image_onset', stop='image_offset')
```

```python
x=[]
y=[]
dur=[]
for i in data[0]['events']['Efix']:
    if i[3]>0 and i[4]>0 and i[2]<500000:
        x.append(i[3])
        y.append(i[4])
        dur.append(i[2])
```

```python
d = pd.Series(dur)
d=d*0.005
```

```python
imagefile = os.path.abspath('image.jpg')

datafile = cbook.get_sample_data(imagefile)
img = imread(datafile)

pyplot.figure(figsize=(16,10))

# pyplot.imshow(img, zorder=0, extent=[0, 1680, 1050, 0])
pyplot.imshow(img, zorder=0, extent=[0, 1680, 1050, 0])
pyplot.plot(x, y,'b',linewidth=1.0)
pyplot.scatter(x,y, s=d, marker='o', color='blue', alpha=0.5)
pyplot.axis([0, 1680, 1050, 0])
pyplot.axis('off')
# pyplot.show()

pyplot.savefig(r'F:\\Timofey\\test_tracker_experiment\\figure.png', bbox_inches='tight')
```

```python
raw_x=data[0]['x']
raw_y=data[0]['y']
```

```python
pyplot.figure(figsize=(16,10))

pyplot.imshow(img, zorder=0, extent=[0, 1680, 1050, 0])
pyplot.plot(raw_x, raw_y, 'y')
pyplot.plot(raw_x, raw_y, 'yo')
pyplot.axis([0, 1680, 1050, 0])
pyplot.show()
```
